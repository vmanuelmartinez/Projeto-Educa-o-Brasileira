1 - Visualizando os dados

```python
import pandas as pd
import matplotlib as plt

#Cores que iremos usar nos gráficos.
colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A6', '#33FFF5', '#FF8C33', '#8C33FF', 
          '#F533FF', '#F5FF33', '#33FFA6', '#FF5733', '#33FF57', '#3357FF', '#FF33A6', 
          '#33FFF5', '#FF8C33', '#8C33FF', '#F533FF', '#F5FF33', '#33FFA6', '#FF5733', 
          '#33FF57', '#3357FF', '#FF33A6', '#33FFF5', '#FF8C33']

dados_educacao = pd.read_excel("Dados-limpos.xlsx")
display(dados_educacao)

```

2 - Analises do % de matri­culas da rede publica no ensino fundamental 2017

```python
import matplotlib.pyplot as plt

territorios = dados_educacao['Territorialidades']
porcentagem = dados_educacao['% de matri­culas da rede publica no ensino fundamental 2017']


plt.figure(figsize=(10, 10))
plt.title('% de Matrículas da Rede Pública no Ensino Fundamental (2017)', color='#1641c6')
plt.barh(territorios, porcentagem, color='silver')
plt.xlabel('Matrículas da Rede Pública no Ensino Fundamental', color='black', fontsize='12')
plt.ylabel('Estados do Brasil', color='black', fontsize='12')

for i in range(len(territorios)):
    plt.text(porcentagem[i] + 0.5, i, f'{porcentagem[i]:.2f}%', 
             va='center', color='black', fontsize=10)
plt.savefig("Gráfico do % de matrículas da rede pública no ensino Fundamental 2017.png")
plt.show()
```

3 - Analise de Analfabetismo de pessoas de 15 anos na Região Norte


```python
import matplotlib.pyplot as plt
RegiaoNorte = ['Acre', 'Amapá', 'Amazonas', 'Pará', 'Rondônia', 'Roraima', 'Tocantins']

analfabetismo15 = [12.60, 18.19, 5.63, 6.66, 12.79, 13.90, 2.94]

plt.figure(figsize=(10, 7))
plt.pie(analfabetismo15, 
        labels=RegiaoNorte, 
        autopct='%1.1f%%', 
       
        colors=plt.cm.Paired(range(len(RegiaoNorte))))

plt.title('Analfabetismo na Região Norte')
plt.xlabel('Colocamos os estados da Região Norte', color='black', fontsize='12')
plt.savefig("Analfabetismo de pessoas de 15 anos na Região Norte.png")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

territorios = ['Acre', 'Alagoas', 'Amapá', 'Amazonas', 'Bahia', 'Ceará', 'Distrito Federal', 'Espírito Santo'
]
homem = [45.82, 39.63, 43.90, 45.33, 40.01, 38.26, 35.56, 42.23]
mulher = [49.83, 43.41, 46.98, 47.53, 43.98, 41.21, 38.05, 45.26]


plt.figure(figsize=(14, 7))
plt.bar(territorios, mulher, width=0.4, label='mulher', color='#9b59b6')
plt.bar(territorios, homem, width=0.4, label='homem', color='#196f3d')

plt.legend()

plt.title("Comparação em desagregação de % de matriculas em homens e mulheres", fontsize='14', color='#a93226')
plt.xlabel("Homems - Mulheres", fontsize=18, color='#a93226')
plt.savefig("Comparação em desagregação de % de matriculas em homens e mulheres.png")
plt.show()
```
